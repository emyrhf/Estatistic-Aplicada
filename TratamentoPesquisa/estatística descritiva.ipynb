## **Medidas de dispersão**

```python
df['altura(m)'].var() # Variância amostral
```

```python
df['altura(m)'].var(ddof=0) #Variância Populacional
```

```python
df['altura(m)'].std() # Desvio padrão amostral
```

```python
# sqrt = raiz quadrada
np.sqrt(df['altura(m)'].var(ddof=0)) # Desvio padrão populacional
```

```python
df['altura(m)'].mean() # Média
```

```python
# Desvio absoluto
(df['altura(m)'] - df['altura(m)'].mean()).abs()
```

```python
# Soma do desvio absoluto
sum((df['altura(m)'] - df['altura(m)'].mean()).abs())
```

```python
df.describe()
```
