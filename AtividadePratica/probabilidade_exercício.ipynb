```python
# CRIANDO FUNÇÃO PROBABILIDADE
def probab (A, E):
  resultado = (A / E)*100
  print('{:.2f}'.format(resultado))
```

```python
# PROBABILIDADE DE RETIRAR PREFERÊNCIA COR PRETO
probab (len(cor_preto), len(dados))
```

## Probabilidade de não ocorrer um evento

```python
# CRIANDO FUNÇÃO PROBABILIDADE DE NÃO OCORRER UM EVENTO
def probab_nao (A, E):
  resultado = (1- (A / E))*100
  print('{:.2f}'.format(resultado))
```

```python
cor_vermelho = dados.loc[dados.cor_favorita.isin(['vermelha','vermelho'])]
cor_vermelho.head()
```

```python
# PROBABILIDADE DE NÃO RETIRAR COR FAVORITA VERMELHO
probab_nao (len(cor_vermelho), len(dados))
```

**EXERCÍCIOS PARA ENTREGAR**

Criar funções para resolver

```python
#FUNÇÕES DAS PROBABILIDADES

#1. PROBABILIDADE CONDICIONAL
def probabEventIndep(A, B, E):
  probA = (A/E)
  probB = (B/E)
  resultado = ((probA*probB)*100)
  print('{:.2f}'.format(resultado))

#2. PROBABILIDADE UNIÃO DE EVENTOS
def probabEventUni(A, B, E):
  probA = (A/E)
  probB = (B/E)
  probIntersec = (probA*probB)
  resultado = (probA + probB - probIntersec)*100
  print('{:.2f}'.format(resultado))

#3. PROBABILIDADE CONDICIONAL
def probabEventCond(A, B, E):
  probA = (A/E)
  probB = (B/E)
  probIntersec = (probA*probB)
  resultado = ((probIntersec/probB)*100)
  print('{:.2f}'.format(resultado))
```

```python
# PROBABILIDADE DE SORTEAR DUAS PESSOAS, COM REPOSIÇÃO, SENDO UMA DELAS COM
# A ALTURA MENOR QUE 1,70M E A OUTRA COR FAVORITA PRETO.
probabEventIndep(len(altura1), len(cor_preto), len(dados))
```

```python
# PROBABILIDADE DE SORTEAR UMA PESSOA QUE TENHA COMO COR FAVORITA PRETA OU
# A ALTURA MENOR QUE 1,70M.
probabEventUni(len(altura1), len(cor_preto), len(dados))
```

```python
# Probabilidade de sortear uma pessoa que tem cor favorita preto dado que (com a condição) tenha altura menor que 1,70
probabEventCond(len(cor_preto), len(altura1), len(dados))
```

```python
# Probabilidade de sortear uma pessoa que tenha altura menor que 1,70 dado que (com a condição) sua cor favorita é preto
probabEventCond(len(altura1), len(cor_preto), len(dados))
```
